14. Save the preprocessing steps in the end model in the pickle-file.

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)
import os
```

```python
train_df = pd.read_csv('train_tweets.csv')
train_df.head()
```

```python
test_df = pd.read_csv('test_tweets.csv')
test_df.head()
```

```python
combine_df = train_df.append(test_df, ignore_index=True, sort=False)
```

```python
combine_df.head()
```

```python
combine_df.info()
```

```python
# 1. remove @user from all tweets usong pattern '@[\w]*' by creating function:
#    - to find every pattern in texst we will use re.findall(pattern, input_txt)
#    - to change @user to blank, we will use re.sub()

def remove_word_user(input_txt):
    pattern = "@[\w]*"
    lst = re.findall(pattern, input_txt)
    result = input_txt
    for el in lst:
        result = re.sub(el, ' ', result)
    return result

combine_df.tweet[1]
```

```python
remove_word_user(combine_df.tweet[1])
```

```python
combine_df['tweet'] = np.vectorize(remove_word_user)(combine_df['tweet'])

combine_df.head()
```

```python
# 2. Use lower register for all tweets with .lower()

combine_df['tweet'] = combine_df['tweet'].apply(lambda x: x.lower())
combine_df.head()
```

```python
# 3. replace apostrophes (for ex "ain't") with empty string using apostrophe_dict. 
#    Creating fucntion: for each word check (for word in text.split()) 
#    if this word is in dictionary (apostrophe_dict)as a key, than change key on it's value.

def replace_by_dict(input_text,my_dict):
    list_of_words = []    
    for word in input_text.split():
        list_of_words.append(my_dict.get(word, word))
    return " ".join(list_of_words)


combine_df['tweet'][7]
```

```python
replace_by_dict(combine_df['tweet'][7], apostrophe_dict)
combine_df['tweet'] = combine_df['tweet'].apply(lambda x: replace_by_dict(x, apostrophe_dict))
combine_df['tweet'][7]
```

```python
# 4. Replace punctuation with empty string, using re.sub(), and pattern r'[^\w\s]'.

combine_df['tweet'] = combine_df['tweet'].apply(lambda x: re.sub(r'[^\w\s]', ' ', x))
combine_df.tail()
```

```python
# 5. Replace abbreviations on its full form using shot_word_dict. Using the same function as above(3.)

combine_df['tweet'] = combine_df['tweet'].apply(lambda x: replace_by_dict(x, short_word_dict))
combine_df.tail()
```

```python
# 6. Replace emoticons with empty string using emoticon_dict. Using the same function as above(3.)

combine_df['tweet'] = combine_df['tweet'].apply(lambda x: replace_by_dict(x, emoticon_dict))
combine_df.head()
```

```python
# 7. Replace special symbols with empty string  using re.sub() and pattern r'[^a-z-Z0-9]'

combine_df['tweet'] = combine_df['tweet'].apply(lambda x : re.sub(r'[^a-zA-Z0-9]', ' ', x))

combine_df.head()
```

```python
# 8. Replace numeral values with empty string using re.sub() and pattern r'[^a-z-A-Z]'
combine_df['tweet'][10]
```

```python
combine_df['tweet'] = combine_df['tweet'].apply(lambda x : re.sub(r'[^a-zA-Z]', ' ', x))

combine_df['tweet'][10]
```

```python
# 9. Remove one-character words using ''.join([w for w in x.split() if len(w)>1])
combine_df['tweet'] = combine_df['tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>1]))
combine_df.head()
```

```python
# 10.Tokenize tweets using nltk.tokenixe.word_tokenize creating new column 'tweet_token'

nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
combine_df['tweet_token'] = combine_df['tweet'].apply(lambda x: word_tokenize(x))
combine_df.head()
```

```python
# 11.Remove stop-words from tokens using nltk.corpus.stopwords. 
#    Create column 'tweet_token_filtered' without stop-words.

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

combine_df['tweet_token_filtered'] = combine_df['tweet_token'].apply(lambda x: [word for word in x if not word in stop_words])
combine_df.head()
```

```python
# 12. Apply stemming(process of finding the root of words) to tokens using nltk.stem.PorterStemmer. 
#     Create olumn ' tweet_stemmed'
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

combine_df['tweet_stemmed'] = combine_df['tweet_token_filtered'].apply(lambda x: [stemmer.stem(word) for word in x])
combine_df.head()
```

```python
# 13. Apply lemmatization (process of finding the form of the related word in the dictionary) to tokens 
#     using nltk.stem.wordnet.WordNetLemmatizer 
#     Create column "tweet_lemmatized"
from nltk.stem import WordNetLemmatizer

nltk.download('omw-1.4')
nltk.download('wordnet')
```

```python
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
combine_df['tweet_lemmatized'] = combine_df['tweet_token_filtered'].apply(lambda x: [lemmatizer.lemmatize(word, nltk.corpus.wordnet.VERB)
 for word in x])
combine_df.head()
```
